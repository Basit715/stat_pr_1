# Q-Q plot
used to check whether the given distribution is normally distributed or not

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
ln = np.random.lognormal(3.0, 1.0, 1000)
```

```python
sns.histplot(ln, kde=True)
```

```python
def plot_data(sample):
     plt.figure(figsize=[10,6])
     plt.subplot(1,2,1)
     sns.histplot(sample,kde=True)
     
     plt.subplot(1,2,2)
     stat.probplot(sample, dist='norm', plot=pylab)
     
     plt.show()
```

```python
plot_data(ln)
```

```python
plot_data(normal_data)
```

# probability distribution function
1. probability mass function(discrete probability function)
     . uniform distribution
     . bernouli distribution
     . binomial distribution
     . poisson distribution

2. probability density function(continous probability function)
     . uniform distribution
     . normal distribution
     . standard normal distribution
     . log normal distribution
     . parrito distribution

```python
# uniform distribution

from numpy import random
uniform = random.uniform(size=500)
sns.distplot(uniform, hist=False)
```

```python
discrete_uniform = np.random.randint(1,8,size=900)
sns.histplot(discrete_uniform)
```

```python
# bernouli distribution

sample_size = 1000
p = 0.3

b_data = np.random.choice([0,1], size=sample_size, p=[1-p,p])

sns.histplot(b_data, bins=2, kde=False, color='red')
```

```python
# binomial distribution
no_of_trials = 10
prob = 0.6
binomial = np.random.binomial(no_of_trials, prob, 1000)

sns.distplot(binomial)
```

```python
poisson_data = np.random.poisson(lam=5, size=1000)
ax = sns.distplot(poisson_data, kde=False, color='red')
ax.set(xlabel='poisson distribution', ylabel='frequency')
plt.show()
```

```python
sample1 = np.random.choice(population, 20)
sample2 = np.random.choice(population, 20)
sample3 = np.random.choice(population, 20)
sample4 = np.random.choice(population, 20)
```

```python
print(sample1)
print(60*'=')
print(sample2)
print(60*'=')
print(sample3)
print(60*'=')
print(sample4)
print(60*'=')
```

```python
mean_of_the_samples = []

all_samples = [sample1,sample2,sample3,sample4]

for i in all_samples:
     mean = np.mean(i)
     mean_of_the_samples.append(mean)

print("mean of all samples =: ",mean_of_the_samples)

```

```python
sns.histplot(mean_of_the_samples, kde=True)
```
